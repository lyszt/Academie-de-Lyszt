```python
from typing import Any, Dict, List

hash_example: Dict[Any,Any] = {}
size: int = 11
for square in range(size):
    hash_example[square] = 0

hash_function = lambda x,y: x % y


def insert_hash(num, size):
    index = hash_function(num, size)
    try:
        print(f"Inserindo {num}, resultado: {index}")
        if hash_example[index] != 0:
            raise KeyError
        hash_example[index] = num
        return index
    except KeyError:
        print(f"Conflito! - Num: {num}, index = {index}")

to_insert: List[int] = [7,17,36,100,106,205]
for i, num in enumerate(to_insert):
    insert_hash(num, size)
    print(hash_example)
```
